```python
import random
import datetime
# import prettytable                  # пакет для таблицы
import matplotlib.pyplot as plt     # библиотека для графика

def BubbleSort(A):                  # сортировка пузырьком
    for i in range(len(A)):
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                a = A[j]
                A[j] = A[j+1]
                A[j+1] = a


def QuickSort(A, fst, lst):         # быстрая сортировка
    if fst >= lst:
        return

    #i, j = fst, lst
    pivot = A[fst]
    # pivot = A[random.randint(fst, lst)]
    first_bigger = fst+1
    while first_bigger <= lst:
        if A[first_bigger] >= pivot:
            break
        first_bigger += 1
    i = first_bigger+1
    while i <= lst:
        if A[i] < pivot:
            A[i], A[first_bigger] = A[first_bigger], A[i]
            first_bigger += 1
        i += 1

    last_smaller = first_bigger-1
    A[fst], A[last_smaller] = A[last_smaller], A[fst]
    QuickSort(A, fst, last_smaller-1)
    QuickSort(A, first_bigger, lst)

def InsertSort(A):                  # сортировка вставками
    for i in range(1, len(A)):
        t = A[i]
        j = i
        while j > 0 and A[j-1] > t:
            A[j] = A[j-1]
            j -= 1
        A[j] = t

def ShakerSort(A):                  # шейкерная сортировка 
    N = len(A)
    for i in range(N // 2):
        for j in range(i, N-1-i):
            if A[j] > A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
        for j in range(N-2-i, i-1, -1):
            if A[j] < A[j-1]:
                A[j], A[j-1] = A[j-1], A[j]


# table = prettytable.PrettyTable(["Размер списка", "Время пузырька", "Время быстрой"])
x=[]
y1=[]
y2=[]
y3 = []
y4 = []



for N in range(1000,5001,1000):
    x.append(N)
    min=1
    max=N
    A=[]
    for i in range (N):
        A.append(int(round(random.random()*(max-min)+min)))

    #print(A)

    B = A.copy()
    # print(B)

    # BubbleSort(A)
    print("---")
    # print(A)



    # QuickSort(B, 0, len(B)-1)
    print("---")
    # print(B)

    t1 = datetime.datetime.now()
    BubbleSort(A)
    t2 = datetime.datetime.now()
    y1.append((t2-t1).total_seconds())
    print("Пузырьковая сортировка   " +str(N)+"   заняла   "+str((t2-t1).total_seconds()) + "c")

    t3 = datetime.datetime.now()
    QuickSort(B, 0, len(B)-1)
    t4 = datetime.datetime.now()
    y2.append((t4 - t3).total_seconds())
    print("Быстрая   " +str(N)+"   заняла   "+str((t4-t3).total_seconds()) + "c")


# Измерение времени сортировки вставками

    C = A.copy()
    t5 = datetime.datetime.now()
    InsertSort(C)
    t6 = datetime.datetime.now()
    y3.append((t6 - t5).total_seconds())
    print("Сортировка вставками   " + str(N) + "   заняла   " + str((t6 - t5).total_seconds()) + "c")

    # Измерение времени шейкерной сортировки
    D = A.copy()
    t7 = datetime.datetime.now()
    ShakerSort(D)
    t8 = datetime.datetime.now()
    y4.append((t8 - t7).total_seconds())
    print("Шейкерная сортировка   " + str(N) + "   заняла   " + str((t8 - t7).total_seconds()) + "c")



    # table.add_row([str(N), str((t2-t1).total_seconds()), str((t4-t3).total_seconds())])
# print(table)

plt.plot(x, y1, "C0", label="Пузырьковая сортировка") 
plt.plot(x, y2, "C1", label="Быстрая сортировка")    
plt.plot(x, y3, "C2", label="Сортировка вставками")  
plt.plot(x, y4, "C3", label="Шейкерная сортировка")

# Добавление заголовка и подписей осей
plt.title("Время выполнения сортировки")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")

plt.legend()

plt.show()

import matplotlib.pyplot as plt

# Точечный график 
plt.figure(figsize=(10, 6))
plt.scatter(x, y1, color="C0", label="Пузырьковая сортировка")
plt.scatter(x, y2, color="C1", label="Быстрая сортировка")
plt.scatter(x, y3, color="C2", label="Сортировка вставками")
plt.scatter(x, y4, color="C3", label="Шейкерная сортировка")
plt.title("Точечный график для сортировок")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")
plt.legend()
plt.show()

# Bar-график
plt.figure(figsize=(12, 6))
width = 150  
plt.bar([i - width*1.5 for i in x], y1, color="C0", width=width, label="Пузырьковая сортировка")
plt.bar([i - width*0.5 for i in x], y2, color="C1", width=width, label="Быстрая сортировка")
plt.bar([i + width*0.5 for i in x], y3, color="C2", width=width, label="Сортировка вставками")
plt.bar([i + width*1.5 for i in x], y4, color="C3", width=width, label="Шейкерная сортировка")

plt.title("Бар-график для сортировок")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")
plt.legend()
plt.show()

# Отдельный график для пузырьковой сортировки
plt.figure(figsize=(10, 6))
plt.plot(x, y1, color="C0", label="Пузырьковая сортировка")
plt.title("Пузырьковая сортировка")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")
plt.legend()
plt.show()

# Отдельный график для быстрой сортировки
plt.figure(figsize=(10, 6))
plt.plot(x, y2, color="C1", label="Быстрая сортировка")
plt.title("Быстрая сортировка")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")
plt.legend()
plt.show()

# Отдельный график для сортировки вставками
plt.figure(figsize=(10, 6))
plt.plot(x, y3, color="C2", label="Сортировка вставками")
plt.title("Сортировка вставками")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")
plt.legend()
plt.show()

# Отдельный график для шейкерной сортировки
plt.figure(figsize=(10, 6))
plt.plot(x, y4, color="C3", label="Шейкерная сортировка")
plt.title("Шейкерная сортировка")
plt.xlabel("Размер списка")
plt.ylabel("Время (сек)")
plt.legend()
plt.show()

```
